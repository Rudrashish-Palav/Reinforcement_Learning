```python
import gymnasium as gym


```

```python
env = gym.make("CartPole-v1", render_mode="human")
observations, info = env.reset()
print(f"Starting Observation: {observations}")

episode_over = False
total_reward = 0

while not episode_over:
    #Sample an action from environments action space
    action = env.action_space.sample()

    #Take the action in the environment
    observation, reward, terminated, truncated, info = env.step(action)

    # Reward +1 for each step the pole stays up
    # Terminated : True if the pole falls too far
    # Truncated : True if the episode is longer than 500 steps

    total_reward += reward
    episode_over = terminated or truncated

print(f"Total reward: {total_reward}")
env.close()

```
